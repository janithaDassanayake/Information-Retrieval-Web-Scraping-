```python
import requests

def download_file():
    print('start downloading')
    link = 'https://en.wikipedia.org/robots.txt'
    r = requests.get(link)
    file = link.split('/')[-1] 

    with open(file,'wb') as new_file:
        new_file.write(r.content)
    print('Download Completed...!')
    print()
    print()

def read_file():
    cwd = os.getcwd()
    with open('robots.txt','r',encoding="utf8") as file:
            content = file.read()
    return content

download_file()
print(read_file())
```

# 4)	Consider the url ‘http://www2.sas.com/proceedings/sugi30/toc.html’.  Extract all the papers names, from 001-30 to 268-30.

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

web_html = urlopen('http://www2.sas.com/proceedings/sugi30/toc.html')
bsoup = BeautifulSoup(web_html, "html.parser")
papers_names = bsoup.findAll('cite')
for papers_name in papers_names:
    print('\n')
    print(papers_name.get_text())
```

# 5)	Consider the url=’http://www.dictionary.com/browse/’ and the words ‘handy’,’whisper’,’lovely’,’scrape’.

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

def get_word_Definitions(word):
    link = "https://www.dictionary.com/browse/" + word + "?s=t"
    webpage = requests.get(link)
    bsoup = BeautifulSoup(webpage.content, 'html.parser')
    output = bsoup.find_all('span', attrs = {'class':'one-click-content'}) 
    return output[0].text
    
Definition_1 = get_word_Definitions('handy')
print(" [1] handy = "+Definition_1)
print( )


Definition_2 = get_word_Definitions('whisper')
print(" [2] whisper = "+Definition_2)
print( )

Definition_3 = get_word_Definitions('lovely')
print(" [3] lovely = "+Definition_3)
print( )

Definition_4 = get_word_Definitions('scrape')
print(" [4] scrape = "+Definition_4)
print()


Definition_Dict = [['handy', Definition_1],['whisper', Definition_2],['lovely', Definition_3], ['scrape', Definition_4]]
dataframe = pd.DataFrame(Definition_Dict,columns=['Word','Definitions'])
dataframe
```
